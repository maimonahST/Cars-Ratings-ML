```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

## Spilt Dataset

```python
# load dataset

df = pd.read_csv('finalDataset.csv')

df.index.name = 'index'
columns = df.columns
print(columns)

df.head()
```

```python
X = df.iloc[:, : -2 ]
y = df.loc[: ,['Rate']]


# Split dataset to 80% for train and 20% for test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=123, train_size = .8)


# train dataset
train = pd.concat([X_train,y_train], axis=1, ignore_index=False )
train.head()


# test dataset
test = pd.concat([X_test,y_test], axis=1, ignore_index=False )
test.head()


```

```python
# save train and test as csv files 
train.to_csv('Train.csv')
test.to_csv('Test.csv')
```

## Build the Model

```python
lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1] # learning rates

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_features=4, max_depth=2, random_state=0)
    gb_clf.fit(X_train, y_train.Rate)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_test, y_test)))
    
# Let's see what the performance was for different learning rates:
```

```python
gb_clf2 = GradientBoostingClassifier(n_estimators=20, learning_rate=0.1, max_features=4, max_depth=2, random_state=0)
gb_clf2.fit(X_train, y_train.Rate)
predictions = gb_clf2.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test.Rate, predictions))

print("Classification Report")
print(classification_report(y_test.Rate, predictions))
```
